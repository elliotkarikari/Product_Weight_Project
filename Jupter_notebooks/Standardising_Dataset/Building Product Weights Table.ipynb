### Decisions

1. If Canned Item has further descriptions 'drained', 'reheated' etc. Remove these words leaving just the canned product.
2. Remove food code to prevent direct linkage back to the nutrient values in the CoFID tables.

Function for removed information after the word canned and its equivalent food code 

```python
def remove_after_canned(food_name, food_code):
    # Create a regular expression that matches the word "canned".
    regex = r'\bcanned\b'
    
    # Use the re.search() function to search for the regular expression in the food_name argument.
    match = re.search(regex, food_name, re.IGNORECASE)
    
    # If the regular expression is found, return the part of the food_name argument before the match. 
    # The `food_name[:match.end()]` expression returns the part of the `food_name` argument before the match. 
    # The '' string is returned because we do not want to return the part of the string after the match.
    if match:
        return food_name[:match.end()], ''
    
    # If the regular expression is not found, return the food_name argument and the food_code argument.
    return food_name, food_code
```

Looking at all altered data

```python
BeanCan['Food Name'], BeanCan['Food Code'] = zip(*BeanCan.apply(lambda x: remove_after_canned(x['Food Name'], x['Food Code']), axis=1))
BeanCan[BeanCan['Food Code'] == '']
```

Function for removed word boiled and all other string after. Also removing its equivalent food code 

```python
def remove_boiled(food_name, food_code):
    # Create a regular expression that matches the word "canned".
    regexb = r'\bboiled\b'
    
    # Use the re.search() function to search for the regular expression in the food_name argument.
    match = re.search(regexb, food_name, re.IGNORECASE)
    
    # If the regular expression is found, return the part of the food_name argument before the match. 
    # The `food_name[:match.end()]` expression returns the part of the `food_name` argument before the match. 
    # The '' string is returned because we do not want to return the part of the string after the match.
    if match:
        return food_name[:match.end()- len(match.group(0))], ''
    
    # If the regular expression is not found, return the food_name argument and the food_code argument.
    return food_name, food_code
```

```python
# Apply the functions to Reduced List 'Food Name' and 'Food Code' columns. Zip Combines Food Name and Food Code
bean_b['Food Name'], bean_b['Food Code'] = zip(*bean_b.apply(lambda x: remove_boiled(x['Food Name'], x['Food Code']), axis=1))
bean_b[bean_b['Food Code'] == '']
```

                                                    Nuts and seeds

```python
GA = Prod_H.loc[Prod_H['Group']== 'GA']
GA
```

                                                    Fish and fish products

```python
JA = Prod_H.loc[Prod_H['Group']== 'JA']
JA
```

**Reduced**

```python
Priority_JA = ['Haddock, flesh only, raw','Hake, flesh only, raw']
```

```python
Filt_JA = JA[JA['Food Name'].isin(Priority_JA)]
Filt_JA
```

```python
Filt_JA['Super Group'] = 'Fish and fish products'
```

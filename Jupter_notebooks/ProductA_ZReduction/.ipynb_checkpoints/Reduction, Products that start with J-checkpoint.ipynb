### Products beginning with a 'J' 

```python
Prod_J = Labelled.loc[Labelled['Food Name'].str.startswith('J')]
Prod_J
```

Products within a Group (provide identification of the food group and food type to which the food belongs)

```python
Prod_J.groupby('Group')['Food Name'].count()
```

## *Reducing products based on its most basic form. Improve as you go on*

                                                 Cereals and cereal products

Biscuits

```python
AM = Prod_J.loc[Prod_J['Group']== 'AM']
AM
```

```python
AM['Super Group'] = 'Cereals and cereal products'
```

Removed Jellabi - AP - Buns and pastries

Puddings

```python
AS = Prod_J.loc[Prod_J['Group']== 'AS']
AS
```

```python
AS['Super Group'] = 'Cereals and cereal products'
```

                                                   Milk and milk products

Puddings and chilled desserts

```python
BR = Prod_J.loc[Prod_J['Group']== 'BR']
BR
```

**Reduced**

```python
Priority_BR = [ 'Jelly, sugar free, made with water']
```

```python
Filt_BR = BR[BR['Food Name'].isin(Priority_BR)]
Filt_BR
```

```python
Filt_BR['Super Group'] = 'Milk and milk products'
```

Removed Jackfruit - DG - Vegetables, general

Removed FA - Fruit, general

Removed JA

Removed JC

                                            Sugars, preserves and snacks

```python
SC = Prod_J.loc[Prod_J['Group']== 'SC']
SC
```

```python
SC['Food Name'].unique()
```

**Reduced**

```python
Filt_SC = SC.loc[SC['Food Name']== 'Jam, reduced sugar']
Filt_SC
```

```python
Filt_SC['Super Group'] = 'Sugars, preserves and snacks'
```

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([AM,AS], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,Filt_BR], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,Filt_SC], axis=0, join='inner')
Red3
```

                                                       Export

```python
Red3.to_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Reduced_J.csv')
```

### Products beginning with a 'J'

Products within a Group (provide identification of the food group and food type to which the food belongs)

```python
Prod_K.groupby('Group')['Food Name'].count()
```

## *Reducing products based on its most basic form. Improve as you go on*

                                                   Milk and milk products

Ice creams

```python
BP = Prod_K.loc[Prod_K['Group']== 'BP']
BP
```

```python
BP['Super Group'] = 'Milk and milk products'
```

                                                        Vegetables

Vegetables, general

```python
DG = Prod_K.loc[Prod_K['Group']== 'DG']
DG
```

**Reduced**

```python
Filt_DG = DG[DG['Food Name'] == 'Kohl rabi, raw']
Filt_DG
```

```python
Filt_DG['Super Group'] = 'Vegetables'
```

Vegetable dishes

```python
DR = Prod_K.loc[Prod_K['Group']== 'DR']
DR
```

**Reduced**

```python
Filt_DR = DR[DR['Food Name'] == 'Khadhi']
Filt_DR
```

```python
Filt_DR['Super Group'] = 'Vegetables'
```

                                                           Fruit

Fruit, general

```python
FA = Prod_K.loc[Prod_K['Group']== 'FA']
FA
```

**Reduced**

```python
remove = ['Kiwi fruit, flesh only, raw, weighed with skin']
```

```python
Filt_FA = FA[~FA['Food Name'].isin(remove)]
Filt_FA
```

```python
Filt_FA['Super Group'] = 'Fruit'
```

                                                Fish and fish products

White Fish

```python
JA = Prod_K.loc[Prod_K['Group']== 'JA']
JA
```

```python
JA['Super Group'] = 'Fish and fish products'
```

Fatty fish

```python
JC = Prod_K.loc[Prod_K['Group']== 'JC']
JC
```

```python
for c in JC['Food Name']:
    print(c)
```

```python
Filt_JC = JC[JC['Food Name'] == 'Kippers, flesh only, grilled']
Filt_JC
```

```python
Filt_JC['Super Group'] = 'Fish and fish products'
```

Removed JR

                                            Meat and meat products

Offal

```python
MG = Prod_K.loc[Prod_K['Group']== 'MG']
MG
```

```python
MG['Food Name'].unique()
```

**Reduced**

```python
Filt_MG = MG.loc[MG['Food Name']== 'Kidney, lamb, raw']
Filt_MG
```

```python
Filt_MG['Super Group'] = 'Meat and meat products'
```

Removed MR

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([BP,Filt_DG], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,Filt_DR], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,Filt_FA], axis=0, join='inner')
Red3
```

```python
Red4 = pd.concat([Red3,JA], axis=0, join='inner')
Red4
```

```python
Red5 = pd.concat([Red4,Filt_JC], axis=0, join='inner')
Red5
```

```python
Red6 = pd.concat([Red5,Filt_MG], axis=0, join='inner')
Red6
```

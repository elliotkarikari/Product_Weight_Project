#### Request Data from Website

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=..."

headers = {}

request_body = {
    "withWeight": True
}

response = requests.get(url, headers=headers, data=request_body)

if response.status_code == 200:
    foods = response.json()#["results"]

    for food in foods:
        print(foods["products"], foods["weight"])
        
else:
    # The request failed, so print the error message
    print(response.status_code, response.reason)
```

```python
import requests

url = "https://api.spoonacular.com/food/products/search"
api_key = "...."  # Replace with your actual API key

# Set the headers with the API key
headers = {
    "apiKey": api_key
}

# Define the query parameters
query_params = {
    "query": "vegan",
    "number": 100,
    "withWeight": True
}

# Send a GET request to the API
response = requests.get(url, headers=headers, params=query_params)

if response.status_code == 200:
    # The request was successful
    data = response.json()

    # Extract the list of products from the response
    products = data["products"]

    for product in products:
        # Print the product information
        print(product["title"], product["weight"])
else:
    # The request failed, so print the error message
    print("Error:", response.status_code, response.reason)
```

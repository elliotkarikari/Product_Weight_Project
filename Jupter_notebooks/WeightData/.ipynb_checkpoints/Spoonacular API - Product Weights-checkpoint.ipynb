                                          Spoonacular API - Product Weights

#### Import Library

```python
import requests
import pandas as pd 
```

#### Request Data from Website

Finding Vegan Food

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=..."

# Send a GET request to the API
response = requests.get(url)
print(response.status_code, response.reason)

if response.status_code == 200:
     # The request was successful
    foods = response.json()

    for food in foods:
        # Print the product information
        print(foods["products"])
        
else:
    # The request failed, so print the error message
    print(response.status_code, response.reason)
```

### Converting Information Into a Pandas DataFrame

```python
spoontab = pd.DataFrame(foods)
spoontab
```

#### Filtering Out Products

```python
spoon_p = spoontab['products']
```

```python
spoon_p = pd.DataFrame(spoon_p)
pd.set_option('display.max_colwidth', None)
spoon_p.iloc[0]
```

##### Title column has lots of information which needs to be separated. Doing this below

```python
titles = spoon_p['products'].apply(lambda x: x['title']).str.split(',', expand=True)# Extract titles from the dictionaries
#spoon_tab = pd.DataFrame({'title': titles})
#spoon_tab
titles
```

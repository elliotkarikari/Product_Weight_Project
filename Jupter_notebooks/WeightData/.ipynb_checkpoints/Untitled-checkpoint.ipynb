                                          Spoonacular API - Product Weights

```python
import requests
import pandas as pd 
```

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=1e0d8e855e304b7a9cb1a5881579f0e1"

headers = {}

request_body = {
    "withWeight": True
}

response = requests.get(url, headers=headers, data=request_body)

if response.status_code == 200:
    foods = response.json()#["results"]

    for food in foods:
        print(foods["products"], foods["weight"])
        
else:
    # The request failed, so print the error message
    print(response.status_code, response.reason)
```

```python
spoontab = pd.DataFrame(foods)
spoontab
```

```python
spoon_p = spoontab['products']
```

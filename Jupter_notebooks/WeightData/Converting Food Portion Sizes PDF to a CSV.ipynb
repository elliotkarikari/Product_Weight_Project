# Food Portion Sizes 

                                        Extracting text from Food Portion Sizes PDF

Install Libraries 

```python
import pandas as pd
import PyPDF2
```

Open, Load in and read PDF

```python
# Load the PDF file
pdf_file = open(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Food_Portion_Sizes.pdf', "rb")
reader = PyPDF2.PdfFileReader(pdf_file)
total_pages = reader.numPages
```

Check number of pages in book

```python
total_pages
```

Go through all pages of pdf. Extract text and append to empty list

```python
extracted_text = []

for page in range(12, 114):
    page_grab = reader.getPage(page)
    text = page_grab.extractText().strip().replace('\n', ' ')  # Replace line breaks with spaces
    text_items = text.split('\n')  # Split the text into separate items
    extracted_text.extend(text_items)  # Extend the list with the items from the current page
```

Extracted text

```python
extracted_text
```

                           I need to do more filtering before converting it to a pandas data frame


In this code, I go through each item in the extracted_text list, split each item by newline ('\n'), and then go through the resulting split items splitting it again by comma (',').

After running this code, the split_text list will contain all the split items from the extracted_text list, where both newlines and commas are used as delimiters. Each split value will be treated as a separate element in the list.

```python
split_text = []

for item in extracted_text:
    split_items = item.split('\n')  # Split each item by newline
    for split_item in split_items:
        split_text.extend(split_item.split(','))
```

                                        Save extracted text as a pandas dataframe

```python
df = pd.DataFrame(split_text)
```

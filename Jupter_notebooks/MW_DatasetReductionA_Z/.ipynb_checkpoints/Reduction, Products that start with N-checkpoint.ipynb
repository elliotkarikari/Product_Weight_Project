###### Import Necessary Libraries 

```python
import pandas as pd
```

Read in Data

```python
Labelled = pd.read_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Labelling2021_watercress.csv')
```

### Products beginning with a 'N' 

```python
Prod_N = Labelled.loc[Labelled['Food Name'].str.startswith('N')]
Prod_N
```

Products within a Group (provide identification of the food group and food type to which the food belongs)

```python
Prod_N.groupby('Group')['Food Name'].count()
```

## *Reducing products based on its most basic form. Improve as you go on*

                                                Cereals and cereal products

Pasta

```python
AD = Prod_N.loc[Prod_N['Group']== 'AD']
AD
```

```python
AD['Food Name'].unique()
```

**Reduced**

```python
Filt_AD = AD[AD['Food Name'] == 'Noodles, egg, dried, raw']
Filt_AD
```

```python
Filt_AD['Super Group'] = 'Cereals and cereal products'
```

                                                        Vegetables

Removed DR - Vegetables, general

                                                           Fruit

Fruit, general

```python
FA = Prod_N.loc[Prod_N['Group']== 'FA']
FA
```

```python
FA['Food Name'].unique()
```

**Reduced**

```python
Filt_FA = FA[FA['Food Name'] == 'Nectarines, flesh and skin, raw']
Filt_FA
```

```python
Filt_FA['Super Group'] = 'Fruit'
```

                                                    Nuts and seeds

Nut and seeds, general 

```python
GA = Prod_N.loc[Prod_N['Group']== 'GA']
GA
```

```python
GA['Super Group'] = 'Nuts and seeds'
```

                                                    Herbs and Spices 

```python
H = Prod_N.loc[Prod_N['Group']== 'H']
H
```

```python
H['Super Group'] = 'Herbs and Spices'
```

                                            Sugars, preserves and snacks

Non chocolate confectionery

```python
SEC = Prod_N.loc[Prod_N['Group']== 'SEC']
SEC
```

```python
SEC['Super Group'] = 'Sugars, preserves and snacks'
```

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([Filt_AD,Filt_FA], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,GA], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,H], axis=0, join='inner')
Red3
```

```python
Red4 = pd.concat([Red3,SEC], axis=0, join='inner')
Red4
```

                                                       Export

```python
Red4.to_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Reduced_N.csv')
```

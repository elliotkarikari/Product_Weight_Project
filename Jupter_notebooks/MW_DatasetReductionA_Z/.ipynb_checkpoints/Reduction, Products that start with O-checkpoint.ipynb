## *Reducing products based on its most basic form. Improve as you go on*

                                                Cereals and cereal products

Biscuits

```python
AM = Prod_O.loc[Prod_O['Group']== 'AM']
AM
```

**Reduced**

```python
Filt_AM = AM[AM['Food Name'] == 'Oatcakes, plain, retail']
Filt_AM
```

```python
Filt_AM['Super Group'] = 'Cereals and cereal products'
```

Removed CD & CDE - Omelettes

                                                        Vegetables

Removed D 

Veg general 

```python
DG = Prod_O.loc[Prod_O['Group']== 'DG']
DG
```

```python
DG['Food Name'].unique()
```

**Reduced**

```python
Priority_DG = ['Okra, raw','Onions, raw']
```

```python
Filt_DG = DG[DG['Food Name'].isin(Priority_DG)]
Filt_DG
```

```python
Filt_DG['Super Group'] = 'Vegetables'
```

Removed DR - Vegetables dishes

Removed WCD

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([Filt_AM,Filt_DG], axis=0, join='inner')
Red
```

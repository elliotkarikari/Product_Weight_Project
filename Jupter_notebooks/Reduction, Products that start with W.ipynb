### Products beginning with a 'W' 

```python
prod_w = Labelled.loc[Labelled['Food Name'].str.startswith('W')]
prod_w
```

Products within a Group (provide identification of the food group and food type to which the food belongs)

```python
prod_w.groupby('Group')['Food Name'].count()
```

## *Reducing products based on its most basic form. Improve as you go on*

                                                Cereals and cereal products

Remoced AA - Flours, grains and starches

Biscuits

```python
am = prod_w.loc[prod_w['Group']== 'AM']
am
```

```python
am['Super Group'] = 'Cereals and cereal products'
```

                                                        Vegetables

Vegetables general 

```python
dg = prod_w.loc[prod_w['Group']== 'DG']
dg
```

**Reduced**

```python
filt_dg = dg[dg['Food Name'] == 'Watercress, raw']
filt_dg
```

```python
filt_dg['Super Group'] = ' Vegetables'
```

                                                       Fruit

```python
fa = prod_w.loc[prod_w['Group']== 'FA']
fa
```

**Reduced**

```python
filt_fa = fa[fa['Food Name'] == 'Whitecurrants, raw']
filt_fa
```

```python
filt_fa['Super Group'] = ' Fruit'
```

                                                    Nuts and seeds

```python
ga = prod_w.loc[prod_w['Group']== 'GA']
ga
```

**Reduced**

```python
filt_ga = ga[ga['Food Name'] == 'Walnuts, kernel only']
filt_ga
```

```python
filt_ga['Super Group'] = 'Nuts and seeds'
```

                                                Fish and fish products

Removed JA - White fish

Fatty fish

```python
jc = prod_w.loc[prod_w['Group']== 'JC']
jc
```

```python
jc['Super Group'] = 'Fish and fish products'
```

Removed JM 

                                               Meat and meat products

Removed MR - Meat dishes

                                                      Alcoholic beverages

Powdered drinks and essences

```python
qe = prod_w.loc[prod_w['Group']== 'QE']
qe
```

```python
qe['Food Name'].unique()
```

**Reduced**

```python
priority_qe = ['Wine, red','Wine, white, medium']
```

```python
filt_qe = qe[qe['Food Name'].isin(priority_qe)]
filt_qe
```

```python
filt_qe['Super Group'] = 'Alcoholic beverages'
```

                                        Soups, sauces and miscellaneous foods

Dairy sauces

```python
wcd = prod_w.loc[prod_w['Group']== 'WCD']
wcd
```

```python
wcd['Food Name'].unique()
```

```python
priority_wcd = ['White sauce, savoury, made with skimmed milk, homemade']
```

**Reduced**

```python
filt_wcd = wcd[wcd['Food Name'].isin(priority_wcd)]
filt_wcd
```

```python
filt_wcd['Super Group'] = 'Soups, sauces and miscellaneous foods'
```

Non-salad sauces

```python
wcn = prod_w.loc[prod_w['Group']== 'WCN']
wcn
```

```python
wcn['Super Group'] = 'Soups, sauces and miscellaneous foods'
```

Miscellaneous foods

```python
wy = prod_w.loc[prod_w['Group']== 'WY']
wy
```

```python
wy['Super Group'] = 'Soups, sauces and miscellaneous foods'
```

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([am, filt_dg], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,filt_fa], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,filt_ga], axis=0, join='inner')
Red3
```

```python
Red4 = pd.concat([Red3,jc], axis=0, join='inner')
Red4
```

```python
Red5 = pd.concat([Red4,filt_qe], axis=0, join='inner')
Red5
```

```python
Red6 = pd.concat([Red5,filt_wcd], axis=0, join='inner')
Red6
```

```python
Red7 = pd.concat([Red6,wcn], axis=0, join='inner')
Red7
```

```python
Red8 = pd.concat([Red7,wy], axis=0, join='inner')
Red8
```

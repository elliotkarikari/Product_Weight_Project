## *Reducing products based on its most basic form. Improve as you go on*

                                                           Eggs

Egg dishes

```python
CD = Prod_Q.loc[Prod_Q['Group']== 'CD']
CD
```

```python
CD['Food Name'].unique()
```

**Reduced**

```python
Priority_CD = ['Quiche, vegetable, retail']
```

```python
Filt_CD = CD[CD['Food Name'].isin(Priority_CD)]
Filt_CD
```

```python
Filt_CD['Super Group'] = 'Eggs'
```

Removed DG

                                                    Vegetables

```python
DR = Prod_Q.loc[Prod_Q['Group']== 'DR']
DR
```

```python
DR['Super Group'] = 'Vegetables'
```

Removed FA

                                                     Nuts and seeds

```python
GA = Prod_Q.loc[Prod_Q['Group']== 'GA']
GA
```

```python
GA['Super Group'] = 'Nuts and seeds'
```

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([Filt_CD,DR], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,GA], axis=0, join='inner')
Red2
```

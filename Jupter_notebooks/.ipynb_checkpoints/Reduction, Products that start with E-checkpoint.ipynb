Read in Data

```python
Labelled = pd.read_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Labelling2021_watercress.csv')
```

### Products beginning with a 'E' 

```python
Prod_E = Labelled.loc[Labelled['Food Name'].str.startswith('E')]
Prod_E.head()
```

Products within a Group (provide identification of the food group and food type to which the food belongs)

```python
Prod_E.groupby('Group')['Food Name'].count()
```

## *Reducing products based on its most basic form. Improve as you go on*

Removed AN - Eccles Cakes - Already have cakes in other reducded tables.

                                                          Eggs

```python
CA = Prod_E.loc[Prod_E['Group']== 'CA']
CA
```

```python
CA['Food Name'].unique()
```

**Reduced**

```python
Priority_CA = ['Eggs, chicken, whole, raw']
```

```python
Filt_CA = CA[CA['Food Name'].isin(Priority_CA)]
Filt_CA
```

```python
Filt_CA['Super Group'] = 'Eggs'
```

Removed CD - Scrambled eggs 

Removed CDE - Egg fu yung

Removed DG - Endive

                                                            Fruit

Fruit, general

```python
FA = Prod_E.loc[Prod_E['Group']== 'FA']
FA
```

```python
FA['Super Group'] = 'Fruit'
```

                                                Fish and fish products

Removed JA - White fish- Used JC - Eel 

Fatty fish

```python
JC = Prod_E.loc[Prod_E['Group']== 'JC']
JC
```

```python
JC['Super Group'] = 'Fish and fish products'
```

                                                Meat and meat products

Meat dishes 

```python
MR = Prod_E.loc[Prod_E['Group']== 'MR']
MR
```

```python
MR['Super Group'] = 'Meat and meat products'
```

                                                       Beverages

Soft drinks/Carbonated drinks

```python
PCA = Prod_E.loc[Prod_E['Group']== 'PCA']
PCA
```

```python
PCA['Super Group'] = 'Beverages'
```

Removed QI - Egg nog

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([Filt_CA,FA], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,JC], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,MR], axis=0, join='inner')
Red3
```

```python
Red4 = pd.concat([Red3,PCA], axis=0, join='inner')
Red4
```

## *Reducing products based on its most basic form. Improve as you go on*

                                                 Cereals and cereal products

Remove AN - Cakes

                                                   Milk and milk products

Ice creams

```python
BP = Prod_I.loc[Prod_I['Group']== 'BP']
BP
```

**Reduced**

```python
Priority_BP = [ 'Ice cream, dairy, vanilla, soft scoop','Ice cream, non dairy, vanilla, soft scoop',]
```

```python
Filt_BP = BP[BP['Food Name'].isin(Priority_BP)]
Filt_BP
```

```python
Filt_BP['Super Group'] = 'Milk and milk products'
```

Removed BR - Puddings and chilled desserts

Sugars, syrups and preserves

```python
S = Prod_I.loc[Prod_I['Group']== 'S']
S
```

```python
Filt_S['Super Group'] = 'Sugars, preserves and snacks'
```

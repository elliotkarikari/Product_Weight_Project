                                          Spoonacular API - Product Weights

#### Import Library

```python
import requests
import pandas as pd 
```

#### Request Data from Website

Finding Vegan Products

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=..."

# Send a GET request to the API
response = requests.get(url)
print(response.status_code, response.reason)

if response.status_code == 200:
     # The request was successful
    foods = response.json()

    for food in foods:
        # Print the product information
        print(foods["products"])
        
else:
    # The request failed, so print the error message
    print(response.status_code, response.reason)
```

### Converting Information Into a Pandas DataFrame

```python
spoontab = pd.DataFrame(foods)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_colwidth',None)
spoontab.drop(['offset','number','totalProducts','processingTimeMs'], axis=1, inplace=True)
```

#### Filtering Out Products

                                                     Products  0 - 99

```python
spoontab
```

##### Title column has lots of information which needs to be separated. Doing this below

```python
spoontab[['id', 'title', 'image']] = spoontab['products'].apply(lambda x: pd.Series([x['id'], x['title'], x['image']]))
spoontab.drop(['products','type'], axis=1, inplace=True)
spoontab
```

I there there are duplicates

```python
spoontab.drop_duplicates(subset='title', inplace=True)
spoontab
```

```python
spoontab.to_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\API Data\spoonacular_vegan.csv')
```

#### Fething the next 100 products 

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=..."

# Send a GET request to the API
response = requests.get(url)
print(response.status_code, response.reason)

if response.status_code == 200:
    # The request was successful
    foods = response.json()

    # Fetching the next 100 products
    offset = 100  # Set the offset to skip the first 100 products
    url_next = f"{url}&offset={offset}"

    response_next = requests.get(url_next)
    if response_next.status_code == 200:
        foods_next = response_next.json()
        
        # Loop through the products in the next set
        for food in foods_next:
            # Print the product information
            print(foods_next["products"])
    
    else:
        # The request for the next set failed, so print the error message
        print(response_next.status_code, response_next.reason)
        
else:
    # The initial request failed, so print the error message
    print(response.status_code, response.reason)
```

                                              Putting data in Pandas Dataframe

                                                     Products 99 - 199

```python
spoontab2 = pd.DataFrame(foods_next)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_colwidth',None)
spoontab2.drop(['offset','number','totalProducts','processingTimeMs'], axis=1, inplace=True)
spoontab2
```

Filtering Database

```python
spoontab2[['id', 'title', 'image']] = spoontab2['products'].apply(lambda x: pd.Series([x['id'], x['title'], x['image']]))
spoontab2.drop(['products','type'], axis=1, inplace=True)
spoontab2
```

```python
spoontab2.drop_duplicates(subset='title', inplace=True)
spoontab2
```

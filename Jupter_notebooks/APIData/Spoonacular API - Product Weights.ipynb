#### Import Library

```python
import requests
import pandas as pd 
```

#### Request Data from Website

Finding Vegan Products

```python
url = "https://api.spoonacular.com/food/products/search?query=vegan&number=100&apiKey=...."

# Send a GET request to the API
response = requests.get(url)
print(response.status_code, response.reason)

if response.status_code == 200:
     # The request was successful
    foods = response.json()

    for food in foods:
        # Print the product information
        print(foods["products"])
        
else:
    # The request failed, so print the error message
    print(response.status_code, response.reason)
```

### Converting Information Into a Pandas DataFrame

```python
spoontab = pd.DataFrame(foods)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_colwidth',None)
spoontab.drop(['offset','number','totalProducts','processingTimeMs'], axis=1, inplace=True)
```

#### Filtering Out Products

```python
spoontab
```

##### Title column has lots of information which needs to be separated. Doing this below

```python
spoontab[['id', 'title', 'image']] = spoontab['products'].apply(lambda x: pd.Series([x['id'], x['title'], x['image']]))
spoontab.drop(['products','type'], axis=1, inplace=True)
spoontab
```

I there there are duplicates

```python
spoontab.drop_duplicates(subset='title', inplace=True)
spoontab
```

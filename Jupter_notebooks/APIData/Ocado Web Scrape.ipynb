```python
import requests
from bs4 import BeautifulSoup

# Get the HTML for the Ocado website
url = 'https://www.ocado.com/search?entry=vegan&amp'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Format the output using prettify()
formatted_output = soup.prettify()

# Print the formatted output
print(formatted_output.strip())
```

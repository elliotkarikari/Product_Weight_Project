```python
# Product Weighting Project

                                         Data Reduction - Food Purchased In Store

###### Import Necessary Libraries 

import pandas as pd

Read in Data

Labelled = pd.read_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Labelling2021_watercress.csv')

### Products beginning with a 'T' 

prod_t = Labelled.loc[Labelled['Food Name'].str.startswith('T')]
prod_t

Products within a Group (provide identification of the food group and food type to which the food belongs)

prod_t.groupby('Group')['Food Name'].count()

## *Reducing products based on its most basic form. Improve as you go on*

                                                Cereals and cereal products

Remoced AA - Flours, grains and starches

af = prod_t.loc[prod_t['Group']== 'AF']
af

af['Super Group'] = 'Cereals and cereal products'

Puddings

AS = prod_t.loc[prod_t['Group']== 'AS']
AS

AS['Food Name'].unique()

**Reduced**

filt_AS = AS[AS['Food Name'] == 'Tarts, bakewell, iced, retail']
filt_AS

filt_AS['Super Group'] = 'Cereals and cereal products'

                                                    Milk and milk products

Yogurts

bn = prod_t.loc[prod_t['Group']== 'BN']
bn

bn['Super Group'] = ' Milk and milk products'

Puddings and chilled desserts

br = prod_t.loc[prod_t['Group']== 'BR']
br

**Reduced**

br['Food Name'].unique()

priority_br = ['Tiramisu', 'Trifle, fruit, retail']

filt_br = br[br['Food Name'].isin(priority_br)]
filt_br

filt_br['Super Group'] = ' Milk and milk products'

                                                        Vegetables

db = prod_t.loc[prod_t['Group']== 'DB']
db

**Reduced**

filt_db = db[db['Food Name'] == 'Tofu, soya bean, steamed']
filt_db

filt_db['Super Group'] = ' Vegetables'

Vegetables, general

dg = prod_t.loc[prod_t['Group']== 'DG']
dg

dg['Food Name'].unique()

**Reduced**

priority_dg = ['Tomatoes, canned, whole contents', 'Tomatoes, standard, raw','Turnip, flesh only, raw']

filt_dg = dg[dg['Food Name'].isin(priority_dg)]
filt_dg

filt_dg['Super Group'] = 'Vegetables'

Removed DR - Vegetable dishes

                                                    Nuts and seeds

ga = prod_t.loc[prod_t['Group']== 'GA']
ga

ga['Super Group'] = 'Nuts and seeds'

                                                    Herbs and spices

h = prod_t.loc[prod_t['Group']== 'H']
h

**Reduced**

priority_h = ['Tarragon, dried, ground','Thyme, dried, ground','Turmeric, ground']

filt_h = h[h['Food Name'].isin(priority_h)]
filt_h

filt_h['Super Group'] = 'Herbs and spices'

                                                Fish and fish products

Molluscs

ja = prod_t.loc[prod_t['Group']== 'JA']
ja

**Reduced**

filt_ja = ja[ja['Food Name'] == 'Tilapia, raw' ]
filt_ja

filt_ja['Super Group'] = 'Fish and fish products'

Remvoed JA - White fish

Fatty fish

jc = prod_t.loc[prod_t['Group']== 'JC']
jc

jc['Food Name'].unique()

**Removed**

priority_jc = ['Trout, brown, raw','Tuna, canned in brine, drained']

filt_jc = jc[jc['Food Name'].isin(priority_jc)]
filt_jc

filt_jc['Super Group'] = 'Fish and fish products'

Removed JR - Fish products and dishes

                                               Meat and meat products

Turkey

mco = prod_t.loc[prod_t['Group']== 'MCO']
mco

mco['Food Name'].unique()

**Reduced**

priority_mco = ['Turkey, whole, raw']

filt_mco = mco[mco['Food Name'].isin(priority_mco)]
filt_mco

filt_mco['Super Group'] = 'Meat and meat products'

Offal

mg = prod_t.loc[prod_t['Group']== 'MG']
mg

mg['Food Name'].unique()

**Reduced**

priority_mg = ['Tongue, lamb, raw','Tripe, dressed, raw']

filt_mg = mg[mg['Food Name'].isin(priority_mg)]
filt_mg

filt_mg['Super Group'] = 'Meat and meat products'

Meat products

mi = prod_t.loc[prod_t['Group']== 'MI']
mi

mi['Super Group'] = 'Meat and meat products'

Removed MR - Meat dishes

                                                        Beverages

Powdered drinks and essences

paa = prod_t.loc[prod_t['Group']== 'PAA']
paa

paa['Super Group'] = 'Beverages'

Removed PAC

Carbonated drinks

pca = prod_t.loc[prod_t['Group']== 'PCA']
pca

pca['Super Group'] = 'Beverages'

Juices

pe = prod_t.loc[prod_t['Group']== 'PE']
pe

pe['Super Group'] = 'Beverages'

                                            Sugars, preserves and snacks

Sugars, syrups and preserves

sc = prod_t.loc[prod_t['Group']== 'SC']
sc

sc['Super Group'] = 'Sugars, preserves and snacks'

Chocolate confectionery

sea = prod_t.loc[prod_t['Group']== 'SEA']
sea

sea['Food Name'].unique()

**Redueced**

priority_sea = ['Topic/Snickers and own brand equivalents','Truffles, mocha, homemade']

filt_sea = sea[sea['Food Name'].isin(priority_sea)]
filt_sea

filt_sea['Super Group'] = 'Sugars, preserves and snacks'

Non-chocolate confectionery

sec = prod_t.loc[prod_t['Group']== 'SEC']
sec

priority_sec = ['Toffees, mixed','Turkish delight, with nuts, homemade']

**Reduced**

filt_sec = sec[sec['Food Name'].isin(priority_sec)]
filt_sec

filt_sec['Super Group'] = 'Sugars, preserves and snacks'

Savoury snacks

sn = prod_t.loc[prod_t['Group']== 'SN']
sn

sn['Super Group'] = 'Sugars, preserves and snacks'

                                        Soups, sauces and miscellaneous foods

Non-salad sauces

wcn = prod_t.loc[prod_t['Group']== 'WCN']
wcn

wcn['Food Name'].unique()

priority_wcn = ['Tartare sauce', 'Tomato ketchup',]

**Reduced**

filt_wcn = wcn[wcn['Food Name'].isin(priority_wcn)]
filt_wcn

filt_wcn['Super Group'] = 'Soups, sauces and miscellaneous foods'

                                              **Join Reduced Tables**
                                              

Red = pd.concat([af, filt_AS], axis=0, join='inner')
Red

Red2 = pd.concat([Red,bn], axis=0, join='inner')
Red2

Red3 = pd.concat([Red2,filt_db], axis=0, join='inner')
Red3

Red4 = pd.concat([Red3,filt_dg], axis=0, join='inner')
Red4

Red5 = pd.concat([Red4,ga], axis=0, join='inner')
Red5

Red6 = pd.concat([Red5,filt_h], axis=0, join='inner')
Red6

Red7 = pd.concat([Red6,filt_ja], axis=0, join='inner')
Red7

Red8 = pd.concat([Red7,filt_jc], axis=0, join='inner')
Red8

Red9 = pd.concat([Red8,filt_mco], axis=0, join='inner')
Red9

Red10 = pd.concat([Red9,filt_mg], axis=0, join='inner')
Red10

Red11 = pd.concat([Red10,mi], axis=0, join='inner')
Red11

Red12 = pd.concat([Red11,paa], axis=0, join='inner')
Red12

Red13 = pd.concat([Red12,pca], axis=0, join='inner')
Red13

Red14 = pd.concat([Red13,pe], axis=0, join='inner')
Red14

Red15 = pd.concat([Red14,sc], axis=0, join='inner')
Red15

Red16 = pd.concat([Red15,filt_sea], axis=0, join='inner')
Red16

Red17 = pd.concat([Red16,filt_sec], axis=0, join='inner')
Red17

Red18 = pd.concat([Red10,sn], axis=0, join='inner')
Red18

Red19 = pd.concat([Red18,filt_wcn], axis=0, join='inner')
Red19

                                                       Export

Red19.to_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\Reduced_T.csv')
```

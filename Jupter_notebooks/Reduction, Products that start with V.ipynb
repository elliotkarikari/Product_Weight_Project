### Products beginning with a 'T'

## *Reducing products based on its most basic form. Improve as you go on*

                                                        Vegetables

Vegetables, general

```python
dg = prod_v.loc[prod_v['Group']== 'DG']
dg
```

```python
dg['Food Name'].unique()
```

**Reduced**

```python
priority_dg = ['Vegetables, mixed, frozen, boiled in unsalted water']
```

```python
filt_dg = dg[dg['Food Name'].isin(priority_dg)]
filt_dg
```

```python
filt_dg['Super Group'] = 'Vegetables'
```

Vegetable dishes

```python
dr = prod_v.loc[prod_v['Group']== 'DR']
dr
```

```python
dr['Food Name'].unique()
```

**Reduced**

```python
priority_dr = ['Vegeburger, retail, grilled','Vegetable stir fry mix, fried in sunflower oil']
```

```python
filt_dr = dr[dr['Food Name'].isin(priority_dr)]
filt_dr
```

```python
filt_dr['Super Group'] = 'Vegetables'
```

                                               Meat and meat products

Removed MAI

```python
mee = prod_v.loc[prod_v['Group']== 'MEE']
mee
```

```python
mee['Food Name'].unique()
```

**Reduced**

```python
priority_mee = ['Venison, meat only, raw',]
```

```python
filt_mee = mee[mee['Food Name'].isin(priority_mee)]
filt_mee
```

```python
filt_mee['Super Group'] = 'Meat and meat products'
```

REmoved MR - Meat dishes

                                                Alcoholic beverages

```python
qg = prod_v.loc[prod_v['Group']== 'QG']
qg
```

```python
qg['Super Group'] = 'Alcoholic beverages'
```

                                        Soups, sauces and miscellaneous foods

Miscellaneous foods

```python
wy = prod_v.loc[prod_v['Group']== 'WY']
wy
```

```python
wy['Super Group'] = 'Soups, sauces and miscellaneous foods'
```

                                              **Join Reduced Tables**
                                              

```python
Red = pd.concat([filt_dg, filt_dr], axis=0, join='inner')
Red
```

```python
Red2 = pd.concat([Red,filt_mee], axis=0, join='inner')
Red2
```

```python
Red3 = pd.concat([Red2,qg], axis=0, join='inner')
Red3
```

```python
Red4 = pd.concat([Red3,wy], axis=0, join='inner')
Red4
```

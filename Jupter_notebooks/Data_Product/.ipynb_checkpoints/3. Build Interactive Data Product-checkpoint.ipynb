## Visualising Food by Super Group using a network Graph

Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re #regular expressions 
from fuzzywuzzy import fuzz
import networkx as nx
import glob
import os
```

Import Dataset

```python
pd.set_option('display.max_rows', None)
Retail = pd.read_csv(r'C:\Users\medekar\Desktop\Product_Weight_Project\Data\Processed\ReducedwithWeights\RedM&Weight.csv', index_col=0)
Retail.head()
```

                                            Creating and Visualising Graph

### Super Groups

```python
%%time
import networkx as nx
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'Retail' with columns 'Super Group' and 'Food Name'
# If you have a different DataFrame or column names, adjust the code accordingly.

# Create a list of unique super groups
super_groups = set(Retail['Super Group'])

# Create a graph
food_graph = nx.Graph()

# Add nodes - unique categories
categories = super_groups
food_graph.add_nodes_from(categories, bipartite=0)

# Generate positions using spring_layout
pos = nx.spring_layout(food_graph, scale=100, seed=42)  # You can adjust the 'scale' parameter as needed

# Create a central node and position it at the center of the layout
center_node = "Product Categories"
pos[center_node] = (0, 0)

# Add edges from 'Super Group' nodes to the center node
for category in categories:
    food_graph.add_edge(category, center_node)

# Draw graph
plt.figure(figsize=(20, 17))

# Draw nodes and labels for categories
nx.draw_networkx_nodes(food_graph, pos, nodelist=categories, node_size=7000, node_color='#e5879d')
nx.draw_networkx_labels(food_graph, pos, labels={category: category for category in categories}, font_size=9, font_family='sans-serif', font_weight='bold', verticalalignment='center')

# Draw the central node separately with different styling
nx.draw_networkx_nodes(food_graph, pos, nodelist=[center_node], node_size=7000, node_color='#42849B', node_shape='o')
nx.draw_networkx_labels(food_graph, pos, labels={center_node: center_node}, font_size=11, font_family='sans-serif', font_weight='bold', verticalalignment='center')

nx.draw_networkx_edges(food_graph, pos, width=1.5, alpha=0.6, edge_color='black')

plt.axis('off')
plt.grid()

# Add title
plt.title('Fruit Food Graph (Super Groups Linked to Center)')

# Add a legend
#plt.legend([center_node], loc='upper left')

plt.show()
```

### Food Products in Super Group

```python
%%time
# Filter the dataframe to include only rows with 'Alcoholic beverages' as the super group
filtered_retail = Retail[Retail['Super Group'] == 'Fruit']#[25:50]

# Create graph
food_graph = nx.Graph()

# Add nodes - unique categories and foods
categories = set(filtered_retail['Super Group'])
foods = filtered_retail['Food Name']#.apply(lambda x: x.split()[0])  # Extract the first word from each food name
food_graph.add_nodes_from(categories, bipartite=0)
food_graph.add_nodes_from(foods, bipartite=1)

# Add weighted edges between foods and categories
for _, row in filtered_retail.iterrows():
    food_graph.add_edge(row['Super Group'], row['Food Name'])

# Generate positions using spring layout
pos = nx.spring_layout(food_graph, iterations=100, k=2.5)

# Draw graph
plt.figure(figsize=(18, 14))
nx.draw_networkx_nodes(food_graph, pos, nodelist=categories, node_size=8700, node_color='#42849B')
nx.draw_networkx_nodes(food_graph, pos, nodelist=foods, node_size=1500, node_color='#e5879d')
nx.draw_networkx_edges(food_graph, pos, width=1.5, alpha=0.6, edge_color='black')
nx.draw_networkx_labels(food_graph, pos, font_size=9, font_family='sans-serif')
plt.axis('off')
plt.grid()

# Add title
plt.title('Alcoholic Beverages Food Graph')

# Save plot to file
# plt.savefig('food_graph.png')
plt.show()
```

```python
%%time
import dash
from dash import dcc, html
import networkx as nx
import pandas as pd

# Sample data to test the app


# Create a graph
food_graph = nx.Graph()

# Add nodes - unique categories and foods
categories = set(Retail['Super Group'])
food_names = set(Retail[Retail['Super Group'] == 'Fruit']['Food Name'])

food_graph.add_nodes_from(categories, bipartite=0)
food_graph.add_nodes_from(food_names, bipartite=1)

# Generate positions using spring_layout
pos = nx.spring_layout(food_graph, scale=100, seed=42)  # You can adjust the 'scale' parameter as needed

# Create a central node and position it at the center of the layout
center_node = "Product Categories"
pos[center_node] = (0, 0)

# Add edges from 'Super Group' nodes to the center node
for category in categories:
    food_graph.add_edge(category, center_node)

# Add edges from 'Food Name' nodes to their corresponding 'Super Group' nodes
for _, row in Retail.iterrows():
    food_graph.add_edge(row['Super Group'], row['Food Name'])

# Create the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.Div([
        # Food Categories linked to Center graph
        dcc.Graph(
            id='food-categories-graph',
            figure={
                'data': [
                    # Nodes for categories
                    {
                        'x': [pos[category][0] for category in categories],
                        'y': [pos[category][1] for category in categories],
                        'text': [category for category in categories],
                        'mode': 'markers+text',
                        'marker': {
                            'size': 30,
                            'color': '#e5879d',
                        },
                        'textfont': {
                            'size': 9,
                            'color': 'black',
                        },
                        'type': 'scatter',
                    },
                    # Central node
                    {
                        'x': [pos[center_node][0]],
                        'y': [pos[center_node][1]],
                        'text': center_node,
                        'mode': 'markers+text',
                        'marker': {
                            'size': 30,
                            'color': '#42849B',
                        },
                        'textfont': {
                            'size': 11,
                            'color': 'black',
                        },
                        'type': 'scatter',
                    },
                ],
                'layout': {
                    'title': 'Food Categories linked to Center',
                    'showlegend': False,
                    'xaxis': {'visible': False},
                    'yaxis': {'visible': False},
                    'hovermode': 'closest',
                }
            }
        )
    ], style={'width': '49%', 'display': 'inline-block'}),

    html.Div([
        # Fruit Food Graph
        dcc.Graph(
            id='fruit-food-graph',
            figure={
                'data': [
                    # Nodes for categories
                    {
                        'x': [pos[category][0] for category in categories],
                        'y': [pos[category][1] for category in categories],
                        'text': [category for category in categories],
                        'mode': 'markers+text',
                        'marker': {
                            'size': 30,
                            'color': '#42849B',
                        },
                        'textfont': {
                            'size': 9,
                            'color': 'black',
                        },
                        'type': 'scatter',
                    },
                    # Nodes for foods
                    {
                        'x': [pos[food_name][0] for food_name in food_names if food_name in pos],
                        'y': [pos[food_name][1] for food_name in food_names if food_name in pos],
                        'text': [food_name for food_name in food_names if food_name in pos],
                        'mode': 'markers+text',
                        'marker': {
                            'size': 15,
                            'color': '#e5879d',
                        },
                        'textfont': {
                            'size': 9,
                            'color': 'black',
                        },
                        'type': 'scatter',
                    },
                ],
                'layout': {
                    'title': 'Fruit Food Graph',
                    'showlegend': False,
                    'xaxis': {'visible': False},
                    'yaxis': {'visible': False},
                    'hovermode': 'closest',
                }
            }
        )
    ], style={'width': '49%', 'display': 'inline-block'})
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

                        Reading In text from bottom half of the 2015 pdf to extract Average weights

```python
%%time

# The path to the PDF file
pdf_path = r"C:\Users\medekar\Desktop\Product_Weight_Project\Data\fruit_and_vegetable_survey_2015_sampling_report.pdf"

# Dictionary to store extracted information from pages 11 to 1291
extracted_data_dict = {}

# Open the PDF file in binary read mode
with open(pdf_path, "rb") as file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfFileReader(file)

    # Extract data from all pages (pages 11 to 1291)
    start_page = 1292
    end_page = min(1331, pdf_reader.numPages)
    for page_num in range(start_page - 1, end_page):
        # Get the current page object
        page = pdf_reader.getPage(page_num)

        # Extract text from the page
        page_text = page.extractText()

        # Store the extracted text in the dictionary
        extracted_data_dict[page_num + 1] = page_text
```

```python
%%time
# Function to display the text on a specific page
def display_text_on_page(extracted_data_dict, page_number):
    if page_number in extracted_data_dict:
        print(f"Text on Page {page_number}:\n{extracted_data_dict[page_number]}")
    else:
        num_pages = len(extracted_data_dict)
        if num_pages > 0:
            print(f"Page number {page_number} not found in the extracted data. Valid page numbers are 1 to {num_pages}")
        else:
            print("No pages found in the extracted data. Make sure you have executed the code to extract data.")

# Test:
page_number_to_display = 1315
display_text_on_page(extracted_data_dict, page_number_to_display)
```

```python
%%time

def extract_average_weight(text):
    # Regular expression pattern to match the "Average weight of" string
    pattern = r"Average\s*weight\s*of\s*(.*)"

    
    matches = re.findall(pattern, text)
    return matches

# List to store extracted average weight strings from all pages
all_extracted_weights = []

# Loop through all pages (pages 1292 to 1330)
for page_num in range(1292, 1331):
    if page_num in extracted_data_dict:
        page_text_to_extract = extracted_data_dict[page_num]
        extracted_weight = extract_average_weight(page_text_to_extract)
        if extracted_weight is not None:
            all_extracted_weights.append(extracted_weight)
        else:
            print(f"No 'Average weight of' pattern found on Page {page_num}.")
    else:
        print(f"Page {page_num} not found in the extracted data.")

# Printing the extracted average weight strings
print("Extracted average weight strings from all pages:")
for idx, weight_string in enumerate(all_extracted_weights, start=1293):
    print(f"Page {idx}:", weight_string)
```

```python
# Flatten the list of lists
flat_extracted_weights = [item for sublist in all_extracted_weights for item in sublist if isinstance(item, str)]

# Splitting the result list by commas
split_result = []
for weight_string in flat_extracted_weights:
    split_result.extend(weight_string.split(','))

split_weights = []
cleaned_average_weights = []

for item in split_result:
    parts = re.split(r'[=:]', item)
    if len(parts) == 2:
        weight = parts[1].strip()
        split_weights.append(weight)
        cleaned_average_weights.append(parts[0].strip())
    else:
        split_weights.append(None)
        cleaned_average_weights.append(item.strip())

# Create a pandas DataFrame
avgW = pd.DataFrame({'Average Weights': cleaned_average_weights, 'Weights': split_weights})
```

                                    Matching avergae weight text to Matched F&V Tables

```python
# Function to calculate similarity score
def calculate_similarity(row):
    return max(fuzz.token_sort_ratio(row['Food Name'], name) for name in avgw['Food Name'])

full['Similarity Score1'] = full.apply(calculate_similarity, axis=1)

# Set a similarity threshold (adjust as needed)
similarity_threshold = 67

# Filter 'full' based on the similarity threshold
filtered_full = full[full['Similarity Score1'] >= similarity_threshold]

for index, row in filtered_full.iterrows():
    # Use fuzzy matching to find matching rows
    matching_rows = avgw[avgw['Food Name'].apply(lambda x: fuzz.token_sort_ratio(x, row['Food Name'])) > similarity_threshold]
    
    if not matching_rows.empty:
        best_match_index = matching_rows['Food Name'].apply(lambda x: fuzz.token_sort_ratio(x, row['Food Name'])).idxmax()
    
        best_match_row = matching_rows.loc[best_match_index]
        #print(best_match_row)
        
        full.at[index, 'Product Weight'] = best_match_row['Average Weights']

full
```
